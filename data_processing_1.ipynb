```python
days=[1,2,3,4,5]
sleeping=[5,7,2,6,8]
playing=[2,7,9,4,1]
eating=[6,7,3,2,1]
working=[5,7,1,9,3]
plt.plot([],color='r',label='sleeping',linewidth=5)#[][] default parameter.mention or do not mention,it has no effects.
plt.plot([],color='g',label='playing',linewidth=5)
plt.plot([],color='k',label='eating',linewidth=5)
plt.plot([],color='y',label='working',linewidth=5)
plt.stackplot(days,sleeping,playing,eating,working,colors=['r','g','k','y'])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("stack plot")
plt.show()
```
